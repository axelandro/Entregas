```python
import pandas as pd

# Leer el archivo CSV
df = pd.read_csv('country_vaccinations.csv')

# Mostrar información sobre el DataFrame
print(df.info())

# Contar el número de veces que aparece cada valor en la columna vaccines
vaccine_counts = df['vaccines'].value_counts()

# Mostrar el resultado
print(vaccine_counts)

# Sumar la columna total_vaccinations para obtener la cantidad total de vacunas aplicadas
total_vaccinations = df['total_vaccinations'].sum()

# Mostrar el resultado
print(total_vaccinations)

# Agrupar el DataFrame por país y calcular la media de la columna total_vaccinations
mean_vaccinations = df.groupby('country')['total_vaccinations'].mean()

# Mostrar el resultado
print(mean_vaccinations)

# Filtrar registros con la fecha deseada
fecha = "2021-01-29"
vac_29_01 = df.loc[df["date"] == fecha]

# Sumar la columna "total_vaccinations"
total_vac_29_01 = vac_29_01["total_vaccinations"].sum()

print(f"La cantidad de vacunas aplicadas el día {fecha} en todo el mundo es: {total_vac_29_01}")

# Crear la columna "diferencias" en el dataframe original
df["diferencias"] = df["daily_vaccinations"] - df["daily_vaccinations_raw"]

# Crear el nuevo dataframe conDiferencias
conDiferencias = df

print(conDiferencias.head())

# Convertir la columna "date" en un objeto de tipo datetime64
df["date"] = pd.to_datetime(df["date"])

# Calcular el período de tiempo
periodo = df["date"].max() - df["date"].min()

print(f"El período de tiempo entre el registro más reciente y el registro más antiguo es: {periodo}")

# Crear una lista con todas las vacunas utilizadas en el dataset
vacunas = list(df["vaccines"].str.split(", ").explode().unique())
vacunas.sort()

# Iterar sobre cada registro y crear la columna "canVac"
for vacuna in vacunas:
    df.loc[df["vaccines"].str.contains(vacuna), "canVac"] = df["total_vaccinations"]

# Crear el nuevo dataframe conCantidad
conCantidad = df

print(conCantidad.head())

# Convertir la columna "date" a datetime64
df["date"] = pd.to_datetime(df["date"])

# Filtrar registros con fecha anterior al 20 de diciembre de 2020
antes20 = df[df["date"] < "2020-12-20"]

# Obtener un dataframe denominado pfizer con todos los registros donde se haya utilizado la vacuna Pfizer. 
pfizer = df[df['vaccines'].str.contains('Pfizer')]

!pip install xlsxwriter

with pd.ExcelWriter('resultadosReto.xlsx', engine='xlsxwriter') as writer:
    conDiferencias.to_excel(writer, sheet_name='conDiferencias')
    conCantidad.to_excel(writer, sheet_name='conCantidad')
    antes20.to_excel(writer, sheet_name='antes20')
    pfizer.to_excel(writer, sheet_name='pfizer')
```
